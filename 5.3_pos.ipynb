```python
def sentences_from_corpus(corpus, fileids = None):
    from nltk.corpus.reader.plaintext import read_blankline_block, concat
    def read_sent_block(stream):
        sents = []
        for para in corpus._para_block_reader(stream):
            sents.extend([s.replace('\n', ' ') for s in corpus._sent_tokenizer.tokenize(para)])
        return sents
    return concat([corpus.CorpusView(path, read_sent_block, encoding=enc)
                   for (path, enc, fileid)
                   in corpus.abspaths(fileids, True, True)])
```

```python
sents = sentences_from_corpus(nltk.corpus.gutenberg, 'milton-paradise.txt')
story = ' '.join(sents)
words = nltk.word_tokenize(story)
```

```python
mp_norm = [word.lower() for word in words if word.isalpha()]
mp_norm
```

```python
mp_tags = nltk.pos_tag(mp_norm, tagset = 'universal')
```

```python
mp_tags
```

```python
mp_nouns = [word[0] for word in mp_tags if word[1] == 'NOUN']
mp_nouns[:10]
```

```python
mp_fd = nltk.FreqDist(mp_nouns)
```

```python
mp_fd.most_common(10)
```
